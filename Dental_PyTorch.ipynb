```python
class CustomDataset(Dataset):
   
   

  def __init__(self, transform=None):
          self.imgs_path = "/root/yolov5/runs/detect/exp/"
          file_list = glob.glob(self.imgs_path + "*.JPG")
          self.data = []     
          for class_path in file_list:
              class_name = class_path.split("/")[-1]
              matching_features = (self.imgs_path + class_name[:-4] + "/stage23_C3_features.npy")
              matching_labels = (self.imgs_path + "labels/" + class_name[:-3] + "txt")
              for img_path in glob.glob(class_path):
                  self.data.append([img_path, matching_labels,matching_features])
              #print(self.data)

        
  def __len__(self):
          return len(self.data)

  def __getitem__(self, idx):
         img_path, matching_labels,matching_features  = self.data[idx]
         
         
         data = pd.read_csv(matching_labels, sep='\s+', header=None,encoding="utf-8",error_bad_lines=False)
         dataFea = np.load(matching_features)
         

         classdetect= []
         count = 0
         
         detection_class = torch.tensor(data.iloc[:,0])
         xCenter =torch.tensor(data.iloc[:,1])   #x_min
         yCenter =torch.tensor(data.iloc[:,2])  #y_min
         Width =torch.tensor(data.iloc[:,3])  #x_max
         Height =torch.tensor(data.iloc[:,4])   #y_max
         detection_score = torch.tensor(data.iloc[:,5])
         feature_vector = torch.from_numpy(dataFea)
         #print(feature_vector)
         
         #classdetect.append(detection_class.numpy())
         #DetectedClass=torch.FloatTensor(classdetect)
        
         
        
         l =[]
         for i in range(len(xCenter)):
               pos =  (xCenter[i], yCenter[i])
               #center = np.array((xCenter[i]+(Width[i]/2), yCenter[i]+(Height[i]/2)))
               l.append(pos)
                  
         centroid = torch.as_tensor(l)
        
         
        
         
         cropped_images =[]
         cropped_imageL = []
        
         imgs = PIL.Image.open(img_path).convert('RGB')
         image_width, image_height = imgs.size
         imgArrays = np.array(imgs)
         
         X = (xCenter*image_width)          #left
         Y = (yCenter*image_height)         #top
         W = (Width*image_width)          #right
         H = (Height*image_height)         #bottom
         cropped_image = np.zeros((image_height, image_width, 3))
         
                        
         for i in range(len(X)):
                 x1, y1, w, h = X[i], Y[i], W[i], H[i]
                 x_start = int(x1 - (w/2))
                 y_start = int(y1 - (h/2))
                 x_end = int(x_start + w)
                 y_end = int(y_start + h)
                 
                 temp = imgArrays[y_start: y_end, x_start: x_end]
                 cropped_image = temp
                 b = cropped_image.ravel()
                 

                 #cropped_imageL.append(cropped_image)
                 dim = (46, 46)
                 
                 # resize image
                 resized = cv2.resize(temp, dim, interpolation = cv2.INTER_AREA)

                 cropped_image_pixels = resized 
                 
                 b = cropped_image_pixels.ravel()
                 #print("Shape of a list: ",np.shape(b))
                 
                 cropped_images.append(b)

                
                 
                 Tensor_cropped_image = torch.as_tensor(np.array(cropped_images))

                              
         return centroid, feature_vector, detection_class, img_path     
```
